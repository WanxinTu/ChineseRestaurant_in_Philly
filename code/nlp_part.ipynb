```python
import pandas as pd
import spacy
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Assuming you have a DataFrame named 'reviews_df'
# Replace 'your_data.csv' with the actual file path or use your own method to load data
reviews_df = df_nlp_filtered.copy()

# Sentiment Analysis:
# Define a function for sentiment analysis using TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Apply the function to the 'review' column
reviews_df['sentiment_score'] = reviews_df['text'].apply(analyze_sentiment)

# Display the result
print(reviews_df[['text', 'sentiment_score']])
```

```python
# Using VADER (Valence Aware Dictionary and sEntiment Reasoner):
nltk.download('vader_lexicon')
# Initialize the VADER sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Define a function for sentiment analysis using VADER
def analyze_sentiment_vader(text):
    sentiment_scores = sid.polarity_scores(text)
    return sentiment_scores['compound']

# Apply the function to the 'review' column
reviews_df['sentiment_score_vader'] = reviews_df['text'].apply(analyze_sentiment_vader)

# Display the result
print(reviews_df[['text', 'sentiment_score_vader']])
```

```python
import matplotlib.pyplot as plt

# Plot sentiment distribution using TextBlob
plt.hist(reviews_df['sentiment_score'], bins=20, edgecolor='black')
plt.title('Sentiment Distribution (TextBlob)')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Reviews')
plt.show()

# Plot sentiment distribution using VADER
plt.hist(reviews_df['sentiment_score_vader'], bins=20, edgecolor='black')
plt.title('Sentiment Distribution (VADER)')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Filter reviews containing the keyword 'service'
service_reviews = reviews_df[reviews_df['text'].str.contains('service|server|Service|waiter|waitress', case=False)]
service_reviews.info()
# Analyze sentiment for service-related reviews
service_reviews['service_sentiment'] = service_reviews['text'].apply(analyze_sentiment)

# Display the result
print(service_reviews[['text', 'service_sentiment']])
```

```python
# Merge DataFrames based on the 'ID' column
merged_df= pd.merge(df_busID_block, mean_byRest_reviews_sentiment, on='business_id', how='outer')
merged_df= pd.merge(merged_df, mean_byRest_service_sentiment, on='business_id', how='outer')
merged_df=pd.merge(merged_df, mean_byRest_price_sentiment, on='business_id', how='outer')

merged_df = merged_df.rename(columns={'mean_x': 'mean_sentiment_all_reviews', 
                                      'std_x': 'std_sentiment_all_reviews',
                                      'count_x':'count_sentiment_all_reviews',
                                      'stderr_x':'stderr_sentiment_all_reviews',
                                      'mean_y': 'mean_sentiment_service_reviews', 
                                      'std_y': 'std_sentiment_service_reviews',
                                      'count_y':'count_sentiment_service_reviews',
                                      'stderr_y':'stderr_sentiment_service_reviews',
                                      'mean': 'mean_sentiment_price_reviews', 
                                      'std': 'std_sentiment_price_reviews',
                                      'count':'count_sentiment_price_reviews',
                                      'stderr':'stderr_sentiment_price_reviews'
                                     })
merged_df['Rank_sentiment_all_reviews'] = merged_df.groupby('block')['mean_sentiment_all_reviews'].rank(ascending=False)
merged_df['Rank_sentiment_service_reviews'] = merged_df.groupby('block')['mean_sentiment_service_reviews'].rank(ascending=False)
merged_df['Rank_sentiment_price_reviews'] = merged_df.groupby('block')['mean_sentiment_price_reviews'].rank(ascending=False)
merged_df
```
